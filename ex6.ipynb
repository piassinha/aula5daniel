```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dic_1 = {
    'Lojas':['A','B','C','D'],
    'Estoque':[500,100,300,600],
    'Chave':['AA','BB','CC','DD']
}
df_estoque = pd.DataFrame(dic_1)
```

```python
df_estoque.head()
```

```python
dic_vendas={
    'Lojas':['A','B','C','D'],   
    'Jan':[100,80,150,50],
    'Fev':[120,90,170,60],
    'Mar':[150,100,200,80],
    'Abril':[180,110,230,90],
    'Mai':[220,190,350,200],
    'Jun':[230,150,2980,120],
    'Jul':[150,170,300,140],
    'Ago':[260,180,310,150],
    'Set':[240,160,290,130],
    'Out':[220,140,270,110],
    'Nov':[400,200,350,190],
    'Dez':[300,350,400,250],
    'Chave':['AA','BB','CC','DD']

}
df_vendas = pd.DataFrame(dic_vendas)
```

```python
df_eccomerce=pd.merge(df_estoque,df_vendas,how='inner',on='Chave')
```

```python
df_eccomerce.head()
```

```python
df_eccomerce.drop('Lojas_y', axis=1,inplace=True)
```

```python
df_eccomerce.head()
```

```python
df_eccomerce.drop('Chave',axis=1,inplace=True)
```

```python
df_eccomerce.head()
```

```python
df_eccomerce=df_eccomerce.T
```

```python
df_eccomerce.head()
```

```python
df_eccomerce.set_index(['Lojas_x'])
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))

axs[0].bar(df_eccomerce['Lojas_x'], df_eccomerce['Estoque'])
axs[0].set_title('Estoque de produtos')
axs[0].set_xlabel('Lojas')
axs[0].set_ylabel('Quantidade de produtos')

axs[0].grid()

axs[1].bar(df_eccomerce['Lojas_x'], df_eccomerce['Jan'])
axs[1].set_title('Histórico de vendas')
axs[1].set_xlabel('Lojas')
axs[1].set_ylabel('Quantidade de produtos')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
dic_vendas= {
    'Lojas': ['A','B','C','D'],
    
    'Jan':[100,80,150,50],
    'Fev':[120,90,170,60],
    'Mar':[150,100,200,80],
    'Abril':[180,110,230,90],
    'Mai':[220,190,350,200],
    'Jun':[230,150,298,120],
    'Jul':[150,170,300,140],
    'Ago':[260,180,310,150],
    'Set':[240,160,290,130],
    'Out':[220,140,270,110],
    'Nov':[400,200,350,190],
    'Dez':[300,350,400,250],
    'Chave':['AA','BB','CC','DD']

}

df_vendas=pd.DataFrame(dic_vendas)

meses= ['Jan','Fev','Mar','Abril','Maio','Jun','Jul','Ago','Set','Out','Nov','Dez']

plt.figure(figsize=(10,6))

for i, loja in enumerate(df_vendas['Lojas']):
    plt.plot(meses, df_vendas.iloc[i,1:-1],label=f'Loja{loja}', marker='o')

plt.title('Histórico de Vendas de Janeiro a Dezembro')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.legend(title="Lojas")
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout
plt.show()
```
