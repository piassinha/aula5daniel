```python
import numpy as np 
from sklearn.linear_model import LinearRegression# importa da biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
Consumo =np.array([ 50, 140, 30, 75,123]).reshape(-1,1)
Demanda=np.array([ 100, 200, 80,100, 160])
```

```python
modelo=LinearRegression()#armazena a regressao linear na variavel modelo
modelo.fit(Consumo, Demanda)#treina o modelo de regressao linear
```

```python
Consumo_teste = np.array([10,120,40,100,200,]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
demanda_previsto = modelo.predict(Consumo_teste)
```

```python
for i in range(len(Consumo_teste)):
 print(f"Vazão: {Consumo_teste[i][0]} m3/min - Nível previsto: {demanda_previsto[i]} metros")
plt.scatter(Consumo_teste, demanda_previsto, color='blue', label='Dados de Treinamento')
plt.plot(Consumo_teste,demanda_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Vazão de Água (m3/min)')
plt.ylabel('Nível do Tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend()#exibe a legenda do grafico
plt.grid(True)#exibe a grade do grafico
plt.show()#plota o grafico
```
