```python
import numpy as np 
from sklearn.linear_model import LinearRegression# importa da biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
A=np.array([50,100,120,300]).reshape(-1,1)
Valor=np.array ([180.000,300.000,375.000,600.000])
```

```python
modelo=LinearRegression()#armazena a regressao linear na variavel modelo
modelo.fit(A, Valor)#treina o modelo de regressao linear
```

```python
Valor_previsto= modelo.predict(A)
```

```python
A_teste = np.array([60,110,140,350]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
Valor_previsto = modelo.predict(A_teste)
```

```python
for i in range(len(A_teste)):
 print(f"Vazão: {A_teste[i][0]} m3/min - Nível previsto: {Valor_previsto[i]} metros")
plt.scatter(A, Valor, color='blue', label='Dados de Treinamento')
plt.plot(A_teste,Valor_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Vazão de Água (m3/min)')
plt.ylabel('Nível do Tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend()#exibe a legenda do grafico
plt.grid(True)#exibe a grade do grafico
plt.show()#plota o grafico
```
