```python
import numpy as np 
from sklearn.linear_model import LinearRegression# importa da biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
Renda =np.array ([1200,5000,8000,10000,30000]).reshape(-1,1)
Gastos =np.array ([750,1500,4000,6000,18000])
```

```python
modelo=LinearRegression()#armazena a regressao linear na variavel modelo
modelo.fit(Renda, Gastos)#treina o modelo de regressao linear
```

```python
Renda_teste = np.array([1000,5100,4000,3500,20000,]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
Gastos_previsto = modelo.predict(Renda_teste)
```

```python
for i in range(len(Renda_teste)):
 print(f"Vazão: {Renda_teste[i][0]} m3/min - Nível previsto: {Gastos_previsto[i]} metros")
plt.scatter(Renda_teste, Gastos_previsto, color='blue', label='Dados de Treinamento')
plt.plot(Renda_teste,Gastos_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Vazão de Água (m3/min)')
plt.ylabel('Nível do Tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend()#exibe a legenda do grafico
plt.grid(True)#exibe a grade do grafico
plt.show()#plota o grafico
```
