```python
import numpy as np 
from sklearn.linear_model import LinearRegression# importa da biblioteca sklearn o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
Q=np.array([100,200,300,400,500,600,700]).reshape(-1,1)#cria o conjunto de dados
T=np.array([30,35,40,45,60,100,150])#conjunto de dados para o nivel
#cria modelo e treina o modelo de regressao linear
modelo=LinearRegression()#armazena a regressao linear na variavel modelo
modelo.fit(Q, T)#treina o modelo de regressao linear
```

```python
Q_teste = np.array([60,110,140,350]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
T_previsto = modelo.predict(Q_teste)
```

```python

Q_teste = np.array([110,220,330,340,410,510,600]).reshape(-1, 1)
# Fazer previsões com base na vazão de água
T_previsto = modelo.predict(Q_teste)
```

```python
for i in range(len(Q_teste)):
 print(f"quantidade de calor: {Q_teste[i][0]} call temp: {T_previsto[i]} metros")
plt.scatter(Q, T, color='blue', label='Dados de Treinamento')
plt.plot(Q_teste,T_previsto, color = 'red',label='Dados de teste')
plt.xlabel('quantidade de calor')
plt.ylabel('temperatura')
plt.title('Relação entre quantidade de calor e atemperatura')
plt.legend()#exibe a legenda do grafico
plt.grid(True)#exibe a grade do grafico
plt.show()#plota o grafico
```
